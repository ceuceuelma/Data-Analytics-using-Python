##4A.1 Binning dengan banyaknya bins tertentu dan panjang masing-masing bin sama

```python
df41 = df_4.copy()
category = ['Cheap','Medium','Expensive']
df41['bin1'] = pd.cut(df41['price'], 3, labels=category, include_lowest=True)
df41
```

```python
bin1_summary = df41.groupby('bin1').agg(row_cnt=('product_code','count'), min_price=('price','min'), max_price=('price','max'))
bin1_summary['bin_interval'] = bin1_summary['max_price'] - bin1_summary['min_price']
bin1_summary
```

```python
df42 = df41.copy()
bins = np.linspace(min(df_4['price']), max(df_4['price']), 4)
category = ['Cheap','Medium','Expensive']
df42['bin2'] = pd.cut(df42['price'], bins, labels=category, include_lowest=True)
df42
```

```python
bins
```

```python
bin2_summary = df42.groupby('bin2').agg(row_cnt=('product_code','count'), min_price=('price','min'), max_price=('price','max'))
bin2_summary['bin_interval'] = bin2_summary['max_price'] - bin2_summary['min_price']
bin2_summary
```

## 4A.2  Binning dengan panjang masing-masing interval tertentu

```python
interval_range
```

```python
bin3_summary = df43.groupby('bin3').agg(row_cnt=('product_code','count'), min_price=('price','min'), max_price=('price','max'))
bin3_summary['bin_interval'] = bin3_summary['max_price'] - bin3_summary['min_price']
bin3_summary
```

#Bagian 4B - CASE STATEMENT

```python
df46 = df45.copy()
df46['price_class'] = np.select(
    [(df46['product_name'].str.lower().str.contains('tea')) & (df46['price']<=2),
     (df46['product_name'].str.lower().str.contains('tea')) & (df46['price'].between(2, 4, inclusive='right')),
     (df46['product_name'].str.lower().str.contains('tea')) & (df46['price']>4),
     (df46['product_name'].str.lower().str.contains('coffee')) & (df46['price']<=3),
     (df46['product_name'].str.lower().str.contains('coffee')) & (df46['price'].between(3, 5, inclusive='right')),
     (df46['product_name'].str.lower().str.contains('coffee')) & (df46['price']>5)],
    ['Cheap Tea', 'Medium Tea', 'Expensive Tea', 'Cheap Coffee', 'Medium Coffee', 'Expensive Coffee'],
    default='Undetermined'
)
df46[(df46['product_name'].str.contains('TEA')) | (df46['product_name'].str.contains('COFFEE'))]
```

```python
price_class_summary = df46.groupby('price_class').agg(row_cnt=('product_code','count'), min_price=('price','min'), max_price=('price','max'))
price_class_summary['price_interval'] = price_class_summary['max_price'] - price_class_summary['min_price']
price_class_summary
```
